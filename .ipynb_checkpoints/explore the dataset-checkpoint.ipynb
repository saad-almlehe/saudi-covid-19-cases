```python
Daily = saudi_cases[(saudi_cases['Daily / Cumulative'] == 'Daily')]
Daily.info()
```

```python
regions = list(Daily['Region'].unique())
date = list(Daily['Date'].unique())
len(date) , len(regions)
```

```python
new_cases = []
new_death = []
new_recoveries =[]
date_new = []
regions_new = []
for i in date:
    for j in regions:
        new_cases.append(Daily[Daily['Region'] == '{}'.format(j)][Daily['Date'] == '{}'.format(i)]\
                         [Daily['Indicator'] == 'Cases']['Cases (person)'].sum())
        new_death.append(Daily[Daily['Region'] == '{}'.format(j)][Daily['Date'] == '{}'.format(i)]\
                         [Daily['Indicator'] == 'Mortalities']['Cases (person)'].sum())
        new_recoveries.append(Daily[Daily['Region'] == '{}'.format(j)][Daily['Date'] == '{}'.format(i)]\
                         [Daily['Indicator'] == 'Recoveries']['Cases (person)'].sum())
        date_new.append(i)
        regions_new.append(j)
```

```python
new_df = pd.DataFrame()
new_df['date'] = date_new
new_df['country'] = 'Saudi Arabia'
new_df['Region'] = regions_new
new_df['New_Cases'] = new_cases
new_df['New_Death'] =new_death
new_df['New_Recoveries']=new_recoveries
new_df
```

```python
new_df['Region'].value_counts()
```

```python
cleand_df = new_df.copy()
```

```python
cleand_df['Total_Cases'] = 0
cleand_df['Total_Deaths'] = 0
cleand_df['Total_Recoveries'] = 0
```

```python
regions = list(cleand_df['Region'].unique())
```

```python
total_Cases = []
total_deaths = []
total_recoveries=[]

for i in regions:
    cases_total = 0
    deaths_total = 0
    recoveries_total =0
    for j in cleand_df.index:
        if(cleand_df['Region'][j] == i):
            cases_total += cleand_df['New_Cases'][j]
            deaths_total += cleand_df['New_Death'][j]
            recoveries_total += cleand_df['New_Recoveries'][j]
            cleand_df['Total_Cases'][j] = cases_total
            cleand_df['Total_Deaths'][j] = deaths_total
            cleand_df['Total_Recoveries'][j] = recoveries_total
            
```
