#####  data set should be sorted by date 

```python
saudi_cases = saudi_cases.sort_values(by='Date')
saudi_cases.head()
```

```python
saudi_cases = saudi_cases[saudi_cases['Region'] != 'Total']
saudi_cases = saudi_cases[saudi_cases['City'] != 'Total']
saudi_cases.head()
```

```python
#### some regions has wrong name 

saudi_cases.Region.value_counts()
```

```python
for i in saudi_cases.index:
    if(saudi_cases['Region'][i] == 'Ar Riyad'):
        saudi_cases['Region'][i] = 'Al-Riyadh'
    elif(saudi_cases['Region'][i] == 'Makkah Al Mukarramah'):
        saudi_cases['Region'][i] = 'Makkah Al-Mokarramah'
    elif(saudi_cases['Region'][i] == 'Al Madinah Al Munawwarah'):
        saudi_cases['Region'][i] = 'Al Madunah'
    elif(saudi_cases['Region'][i] == 'Al Qaseem'): 
        saudi_cases['Region'][i] = 'Al-Qaseem'
    elif(saudi_cases['Region'][i] == 'Al Jawf'): 
        saudi_cases['Region'][i] = 'Al-Jouf'
    elif(saudi_cases['Region'][i] == 'Tabuk'):
        saudi_cases['Region'][i] = 'Tabouk'
    elif(saudi_cases['Region'][i] == 'Al Bahah'):
        saudi_cases['Region'][i] = 'Al-Baha'
    elif(saudi_cases['Region'][i] == 'Northern Borders'):
        saudi_cases['Region'][i] = 'Al Ḥudud ash Shamaliyah'
```
